```python
# Define a function to print out the columns with nan greater than given threshold.
def printNumNanColumns(threshold):
    column_with_nan = telecom_df.columns[telecom_df.isnull().any()]
    count_nan_columns = 0
    for column in column_with_nan:
        if telecom_df[column].isnull().sum()*100.0/len(telecom_df.index) > threshold:
            print(column, round(telecom_df[column].isnull().sum()*100.0/len(telecom_df.index),2))
            count_nan_columns += 1
        
    print("\nNumber of Columns with nan are {0} for threshold {1}".format(count_nan_columns,threshold))
```

#### But as per the requirement, we need to predict churn only for the high-value customers. i.e., for those who have recharged with an amount more than or equal to 'the 70th percentile of the average recharge amount in the first two months'

So we should not drop the below columns related to the recharge amounts. In order to do so, we need to impute the missing values in these columns -  

- total_rech_data_6
- total_rech_data_7

- av_rech_amt_data_7

And in order to impute the above columns, we have to make the cells as 0 when 'total_rech_data_%' as well as 
'date_of_last_rech_data_%' are both zero

```python
if telecom_df['total_rech_data_6'].isnull().equals(telecom_df['date_of_last_rech_data_6'].isnull()):
    print('NULL Value index for Month6 for total_tech_data and date_of_last_rech_data are equal. Imputing as 0')
    telecom_df['total_rech_data_6'].fillna(0, inplace=True)
    telecom_df['av_rech_amt_data_6'].fillna(0, inplace=True)

if telecom_df['total_rech_data_7'].isnull().equals(telecom_df['date_of_last_rech_data_7'].isnull()):
    print('NULL Value index for Month7 for total_tech_data and date_of_last_rech_data are equal. Imputing as 0')
    telecom_df['total_rech_data_7'].fillna(0, inplace=True)
    telecom_df['av_rech_amt_data_7'].fillna(0, inplace=True)

if telecom_df['total_rech_data_8'].isnull().equals(telecom_df['date_of_last_rech_data_8'].isnull()):
    print('NULL Value index for Month8 for total_tech_data and date_of_last_rech_data are equal. Imputing as 0')
    telecom_df['total_rech_data_8'].fillna(0, inplace=True)
    telecom_df['av_rech_amt_data_8'].fillna(0, inplace=True)
```

#### 2. Tag churners and remove attributes of the churn phase

#### Now we need to tag the churned customers (churn=1, else 0) based on the fourth month as follows: 
#### Those who have not made any calls (either incoming or outgoing) AND have not used mobile internet even once in the churn phase. The attributes we need to use to tag churners are:

- total_ic_mou_9
- total_og_mou_9
- vol_2g_mb_9
- vol_3g_mb_9

```python
# Tag Churners
# Any customer who does not use any of the facility (calls or mobile data) during the 9th month,
# will have the row sum for the attributes equal to zero. 
# This customer can be tagged as Churn (1) else the customer will be tagged as Not-Churn (0).
churn_agg = telecom_df["total_ic_mou_9"] + telecom_df["total_og_mou_9"] + \
                            telecom_df["vol_2g_mb_9"] + telecom_df["vol_3g_mb_9"]
telecom_df["churn"] = np.where(churn_agg, 0, 1)
telecom_df['churn'].head()
```

### Check if we have major outliers present in the data using Quantiles

```python
colNum = 0
for col in telecom_df.describe().columns:
    q75 = round(telecom_df[col].quantile(0.75),1)
    q25 = round(telecom_df[col].quantile(0.25),1)
    minV = round(telecom_df[col].min(),1)
    maxV = round(telecom_df[col].max(),1)
    q50 = round(telecom_df[col].quantile(0.50),1)
    lLimit = round(q25 - (3 * (q75 - q25)),1)
    uLimit = round(q75 + (3 * (q75 - q25)),1)

    if ( (lLimit < q25) & (uLimit > q75) ):
        outlier = telecom_df[ (telecom_df[col] < lLimit) | (telecom_df[col] > uLimit)].shape[0]

        colNum = colNum + 1
        print ('col {0}, minV {1}, maxV {2}, q25 {3}, q50 {4}, q75 {5},lLimit {6}, uLimit {7}, outlier {8}' . format(\
               col, minV, maxV, q25, q50, q75, lLimit, uLimit, outlier))
        
print ("Total columns {0}".format (colNum))
```

### Now apply PCA algorithm in order to scale down the number of variables and also removing any
### Multicolinearity

```python
from sklearn.decomposition import PCA
pca = PCA(svd_solver='randomized', random_state = 42)
```

```python
pca.fit(X_train_scaled)
```

```python
pca.components_.shape
```

As above we see that PCA algorithm by default takes 158 pca componenets which are too big.
Lets analyse the variance ration and see the optimal values of the components with which we can have 
the maximum data variance captured

```python
pca.explained_variance_ratio_
```

```python
%matplotlib inline
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("number pca components")
plt.ylabel("cumulative variance explained by the number of pca components")
plt.show()
```

We see that optimal value lies around 20 PCA componenets which captures above 90 data variance, which should be ok for our model
Lets run the PCA algorithm, providing it a convergence variance of 90%.

```python
colnames = list(X_train_scaled.columns)
pcs_df = pd.DataFrame({'PC1':pca.components_[0],'PC2':pca.components_[1], 'Feature':colnames})
pcs_df.head()
```

```python
%matplotlib inline
fig = plt.figure(figsize = (8,8))
plt.scatter(pcs_df.PC1, pcs_df.PC2)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
for i, txt in enumerate(pcs_df.Feature):
    plt.annotate(txt, (pcs_df.PC1[i],pcs_df.PC2[i]))
plt.tight_layout()
plt.show()
```

```python
# As seen above 20 components can handle more than 90% data variance. lets build a model with components as 20
from sklearn.decomposition import IncrementalPCA
pca_final = IncrementalPCA(n_components=20)
X_train_pca = pd.DataFrame(pca_final.fit_transform(X_train_scaled))
X_train_pca.head()
```

```python
#creating correlation matrix for the principal components
corrmat = np.corrcoef(X_train_pca.transpose())
#plotting the correlation matrix
%matplotlib inline
plt.figure(figsize = (20,10))
sns.heatmap(corrmat,annot = True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

def createOptimalCutOffGraph(pred_df):
    # Draw a Graph for sesitivity and specificity for probability cut off values like
    # 0, 0.1, 0.2 to 0.9
    cutOffval = [float(x)/10 for x in range(10)]
    for val in cutOffval:
        pred_df[val] = pred_df['churnProb'].apply(lambda x: 1 if x > val else 0)

    cutoff_df = pd.DataFrame( columns = ['prob','accuracy','sensi','speci'])
    
    for val in cutOffval:
        cm = metrics.confusion_matrix(pred_df['churn'], pred_df[val] )
        total = sum(sum(cm))
        
        accuracy = (cm[0,0]+cm[1,1])/total
        speci = cm[0,0]/(cm[0,0]+cm[0,1])
        sensi = cm[1,1]/(cm[1,0]+cm[1,1])
        cutoff_df.loc[val] =[ val ,accuracy,sensi,speci]
        
    cutoff_df.plot.line(x='prob', y=['accuracy','sensi','speci'])
    plt.show()
```

```python
createOptimalCutOffGraph(y_pred_final_df)
```

```python
# Function to calculate the confusion metrics and evaluation params
def calculateConfusionMetrics(y_test, y_pred_dt):
    cm = metrics.confusion_matrix(y_test, y_pred_dt )
    total = sum(sum(cm))

    accuracy = (cm[0,0]+cm[1,1])/total
    speci = cm[0,0]/(cm[0,0]+cm[0,1])
    sensi = cm[1,1]/(cm[1,0]+cm[1,1])
    
    print ("Accuracy {0}, Specificity {1}, Sensitivity {2}".format (accuracy, speci, sensi))
    print ("Confusion Matrix ")
    print (cm)
```

```python
# Draw ROC Graph
def createRocGraph(actual, probs):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    
    auc_score = metrics.roc_auc_score( actual, probs )
    
    print ("Auc Score {0}".format(auc_score))
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```
